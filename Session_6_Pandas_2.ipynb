## Group By

```python
data = {'Brand':['iphone','iphone','Howavei','Howavei','Samsung','Samsung'],
        'Person':['Ahmad','Reza','Hasan','Sarah','Majid','Maryam'],
        'Price':[1500,1200,1340,1500,2000,3500]
        
}
```

```python


df = pd.DataFrame(data)
```

```python
df
```

```python
df.groupby('Brand')
```

```python
df.groupby('Brand').sum()
```

```python
df.groupby('Brand').mean('Price')
```

```python
df.groupby('Brand').std()
```

```python
df.groupby(by ='Brand').min()
```

```python
df.groupby(by ='Brand').count()
```

```python
df.groupby('Brand', sort=False).mean()
```

```python
df.groupby('Brand', sort=False).mean().loc['iphone']
```

```python
df.groupby('Brand', sort=False).agg({'sum','count', 'max'})
```

```python
df.groupby('Brand', sort=False).agg({'Price':['sum', 'count', 'max'], 'Person': 'count'})
```

```python
df.groupby('Brand', sort=False).describe()
```

```python
df
```

```python
df.groupby('Brand', sort=False).describe().loc['Howavei']
```

## Operations on DataFrames

#### Merging,Joining,Concatenating

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                    'B': ['B0', 'B1', 'B2', 'B3'],
                    'C': ['C0', 'C1', 'C2', 'C3'],
                    'D': ['D0', 'D1', 'D2', 'D3']},
                   index=[0, 1, 2, 3])
```

```python
df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                   'B': ['B4', 'B5', 'B6', 'B7'],
                   'C': ['C4', 'C5', 'C6', 'C7'],
                   'D': ['D4', 'D5', 'D6', 'D7']},
                index=[4, 5, 6, 7])
```

```python
df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                    'B': ['B8', 'B9', 'B10', 'B11'],
                    'C': ['C8', 'C9', 'C10', 'C11'],
                    'D': ['D8', 'D9', 'D10', 'D11']},
                   index=[8, 9, 10, 11])
```

```python
df1
```

```python
df2
```

```python
df3
```

#### Concatenation

```python
pd.concat([df1,df2,df3])
```

```python
pd.concat([df1,df2,df3], axis = 1)
```

```python
pd.concat([df1,df2], axis = 1)
```

```python
# append()
df1.append(df2)
```

#### Merging

```python
left = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})
```

```python
right = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                      'C': ['C0', 'C1', 'C2', 'C3'],
                      'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
left
```

```python
right
```

```python
pd.concat([left,right], join='inner', axis = 1)
```

```python
pd.merge(left,right)
```

```python
pd.merge(left,right, on = 'key')
```

```python
pd.merge(left,right, how = 'inner', on = 'key')
```

```python
pd.merge(left,right, how = 'outer', on = 'key')
```

```python
left = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                      'key2': ['K0', 'K1', 'K0', 'K1'],
                      'A': ['A0', 'A1', 'A2', 'A3'],
                      'B': ['B0', 'B1', 'B2', 'B3']})
```

```python
right = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                       'key2': ['K0', 'K0', 'K0', 'K0'],
                       'C': ['C0', 'C1', 'C2', 'C3'],
                       'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
left
```

```python
right
```

```python
pd.merge(left,right,on=['key1','key2'],how='inner')
```

```python
pd.merge(left,right,on=['key1','key2'],how='left')
```

```python
pd.merge(left,right,on=['key1','key2'],how='right')
```

```python
pd.merge(left,right,on=['key1','key2'],how='outer')
```

## Pivot Table

```python
df = pd.DataFrame({"A": ["foo", "foo", "foo", "foo", "foo",
                         "bar", "bar", "bar", "bar"],
                   "B": ["one", "one", "one", "two", "two",
                         "one", "one", "two", "two"],
                   "C": ["small", "large", "large", "small",
                         "small", "large", "small", "small",
                         "large"],
                   "D": [1, 2, 2, 3, 3, 4, 5, 6, 7],
                   "E": [2, 4, 5, 5, 6, 6, 8, 9, 9]})
```

```python
df
```

```python
df.value_counts()
```

```python
df.value_counts('A')
```

```python
df.pivot_table(values = 'D', index = ['A','B'], columns = 'C')
```

```python
df.pivot_table(values = 'E', index = ['A','B'], columns = 'C')
```

```python
df.pivot_table(values = 'E', index = ['A'], columns = 'C')
```
